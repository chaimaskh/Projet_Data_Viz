```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger votre jeu de donnÃ©es, par exemple :
# df = pd.read_csv("votre_fichier.csv")

# Calculer le nombre de donnÃ©es manquantes par colonne
missing_data = df.isnull().sum()

# Calculer le pourcentage de donnÃ©es manquantes par colonne
percentage_missing = (missing_data / len(df)) * 100

# CrÃ©er un DataFrame pour stocker les pourcentages manquants par colonne
missing_df = pd.DataFrame({'Colonne': percentage_missing.index, 'Pourcentage de DonnÃ©es Manquantes': percentage_missing.values})

# Trier le DataFrame par pourcentage de donnÃ©es manquantes
missing_df = missing_df.sort_values(by='Pourcentage de DonnÃ©es Manquantes', ascending=False)

# Afficher le graphique des pourcentages manquants par colonne
plt.figure(figsize=(12, 6))
plt.barh(missing_df['Colonne'], missing_df['Pourcentage de DonnÃ©es Manquantes'])
plt.xlabel('Pourcentage de DonnÃ©es Manquantes')
plt.ylabel('Colonne')
plt.title('Pourcentage de DonnÃ©es Manquantes par Colonne')
plt.show()

# Afficher le DataFrame des pourcentages manquants
print(missing_df)
```
